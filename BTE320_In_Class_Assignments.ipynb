Calculate Gross Profit

      Assume the COGS is 40 % of sales

Write a program to:

prompt for Sales
Calculate and display the Cost of Goods sold and Profit

Inputs :

Sales 12000
COGS 4800
Gross profit 7200

Calculate Taxes and net income
  Prompt for tax rate in percentage
  Calculate taxes ( gross profits x tax rate) and net income (gross - taxes)

Display income statement



```python
Sales = float(input("Enter total sales "))
cogs = 0.4 * Sales
Gross_Profit = Sales - cogs
Tax_rate = float(input("Enter tax rate: "))
Taxes = Gross_Profit * Tax_rate
Net_income = Gross_Profit - Taxes

print(f" Sales: $ {Sales}")
print(f" Cost of Goods Sold: ${cogs}")
print(f" Gross Profit: ${Gross_Profit}")
print(f" Taxes : $ {Taxes}")
print(f" Net Income: {Net_income}")
```

```python
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
operator = input("Enter the operator:")
if operator == "*":
    result = num1 * num2
elif operator == "-":
    result = num1 - num2
elif operator == "+":
    result = num1 + num2
elif operator == "/":
    if num2 == 0:
        print("Error: Cannot divide by zero")
        exit()
    result = num1 / num2
else:
    print("Invalid operator.")
    exit()

print(f"The result is: {result}")
```
